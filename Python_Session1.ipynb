# Data types
Python has the following main built-in data types:
Numeric Types → int, float, complex
Text Type → str
Sequence Types → list, tuple, range
Mapping Type → dict
Set Types → set, frozenset
Boolean Type → bool
None Type → NoneType
Complex literals can be created by using the notation x + yj where x is the real component and y is the imaginary component.
The type() function is used to check the data type of a variable.

```python
x = 10          # int
y = 3.14        # float
z = 2 + 3j      # complex

print(type(x))
print(type(y))
print(type(z))
```

```python
name = "Alice"
message = 'Hello World'

print(type(name))
print(type(message))
```

```python
is_active = True
x = None

print(type(is_active))
print(type(x))
```

```python
#complex numbers: note the use of `j` to specify the imaginary part

j = 1.0 - 2.0j
j
print(type(j))
print(j.real, j.imag)
```

```python
#Operators in Python

a = 100
b = 35
print(a + b)   
print(a - b)   
print(a * b)   
print(a / b)  
print(a % b)  
print(a ** b)  
print(a // b) #floor division
```

```python
#comparison operations

a = 5
b = 7
print(a == b)  
print(a != b)  
print(a < b)   
print(a > b)   
print(a <= b) 
print(a >= b)  
```

```python
#logical operations 

x = True
y = False
print(x and y)  
print(x or y)   
print(not y) 
```

```python
#Type Casting
a = int(3.5)
b = float(5)
c = str(100)

print(a)
print(b)
print(c)
```

#control statements
# Conditional Statements – Decide which block of code to execute

# if, elif, else

# Looping Statements – Repeat a block of code multiple times

# for loops, while loops

# Control Flow Modifiers – Alter the flow inside loops

# break → exit loop

# continue → skip current iteration

# pass → placeholder / do nothing


```python
age = 20

if age >= 18:
    print("Adult")
elif age >= 13:
    print("Teenager")
else:
    print("Child")
```

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

```python
for i in range(10):
    if i == 5:
        break   # Exit the loop when i == 5
    print("Break Example:", i)

for i in range(5):
    if i == 2:
        continue  # Skip iteration when i == 2
    print("Continue Example:", i)
```

```python
count = 0

while count < 5:
    print("Count:", count)
    count += 1
```
