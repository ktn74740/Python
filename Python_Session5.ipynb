An iterable is any Python object that can return its elements one at a time.

Examples of iterables:

list

tuple

string

set

dictionary

range

An object is iterable if it implements the special method:
An iterator is an object that:

Keeps track of its current position

Returns the next value using the method:

```python
# Iterable example
nums = [10, 20, 30]
for x in nums:
    print(x)

# String is also iterable
for ch in "hello":
    print(ch)
```

```python
#Converting Iterable to Iterator
numbers = [1, 2, 3, 4]

iterator_obj = iter(numbers)

print(iterator_obj)
```

```python
#Using next()
numbers = [1, 2, 3]

iterator_obj = iter(numbers)

print(next(iterator_obj))
print(next(iterator_obj))
print(next(iterator_obj))
```

```python
#topIteration Error
numbers = [1, 2]

iterator_obj = iter(numbers)

print(next(iterator_obj))
print(next(iterator_obj))
print(next(iterator_obj))  # This will raise StopIteration
```

```python
# Generator function: cubes (power of 3)
def gencubes(n):
    for num in range(n):
        yield num**3

g = gencubes(5)
g  # generator object
```

```python

#Generator Function Using yield
def square_numbers(n):
    for i in range(n):
        yield i * i

gen = square_numbers(5)

print(gen)
```

```python
#Using next() with Generator
gen = square_numbers(5)

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# Normal Fibonacci function that returns a list
def fibon_list(n):
    a, b = 1, 1
    output = []
    for _ in range(n):
        output.append(a)
        a, b = b, a + b
    return output

fibon_list(10)
```

```python
# Fibonacci generator
def genfibon(n):
    """Generate Fibonacci sequence up to n terms"""
    a, b = 1, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in genfibon(12):
    print(num)
```

The map() function applies a given function to each element of an iterable (like a list, tuple, etc.) and 
returns a map object (iterator).
map() returns an iterator

It processes elements one by one

It is memory efficient

Often used with lambda functions

Useful in data preprocessing

```python
#Basic Example
def square(x):
    return x * x

numbers = [1, 2, 3, 4]

result = map(square, numbers)

print(list(result))
```

```python
def fahrenheit(T):
    return (float(9) / 5) * T + 32

def celsius(T):
    return (float(5) / 9) * (T - 32)

temp_f = [0, 22.5, 40, 100]

# Convert F -> C
temp_c = list(map(celsius, temp_f))
temp_c
```

```python
# lambda with map
# A lambda is a small one-time function.

lst = [1, 2, 3, 4, 5]
list(map(lambda x: x + 1, lst))
```

```python
# with multiple iterables
# map() can take multiple iterables if your function takes multiple arguments.

a = [1, 2, 3, 4]
b = [5, 6, 7, 8]

list(map(lambda x, y: x + y, a, b))
```

```python
#  reduce()
# reduce(function, iterable) repeatedly applies a function to produce one final value.

# You must import it:

# from functools import reduce
# Mental model
# If the list is [s1, s2, s3, s4], then reduce does:

# f(s1, s2) → result1
# f(result1, s3) → result2
# f(result2, s4) → final
from functools import reduce

names = ['kota', 'ruchik', 'joey', 'tribiani']
reduce(lambda a, b: a + b, names)
```
