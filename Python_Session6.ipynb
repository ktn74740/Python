Common Types of Errors
Error Type	Description	Example
SyntaxError	Code structure is invalid	print("Hello"
NameError	Using undefined variable	print(x)
TypeError	Invalid operation between types	5 + "5"
ZeroDivisionError	Division by zero	10 / 0
IndexError	Invalid index in list/tuple	[1,2][5]
KeyError	Invalid key in dictionary	d = {}; d['a']

#try-Except Block
try:
    # Code that might raise an exception
except ExceptionType:
    # Code to handle the exception

```python
#Example
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
#Multiple Exceptions
try:
    a = int("abc")
    b = 10 / 0
except ValueError:
    print("Invalid conversion!")
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
#Catch All Exceptions
try:
    x = 10 / 0
except Exception as e:
    print("An error occurred:", e)
```

```python
#Else and Finally
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid input!")
else:
    print("You entered:", num)
finally:
    print("This always runs, whether error occurs or not")
```

```python
#File Operations
# Create a sample text file using write mode (overwrites)
with open("notes.txt", "w", encoding="utf-8") as f:
    f.write("  line one  \n")
    f.write("line two\n")
    f.write("   line three\n\n")

print("Created notes.txt")
```

```python
# Read the file and use strip() to clean lines
with open("notes.txt", "r", encoding="utf-8") as f:
    for line in f:
        cleaned = line.strip()
        if cleaned:  # skip empty lines after stripping
            print(cleaned)
```

```python
# Append mode adds content without overwriting
with open("notes.txt", "a", encoding="utf-8") as f:
    f.write("appended line\n")

print("Appended to notes.txt")
```

 Working with Excel .xlsx
Two common ways:

Pandas (best for analysis) — read_excel, to_excel
openpyxl (best for cell-level editing/formatting)
If you get an engine error, install: pip install openpyxl

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Ritesh", "Nikhil", "Phenome", "Rahul"],
    "score": [95, 88, 91, 77]
})

df
```

```python
# Write to Excel
df.to_excel("students.xlsx", index=False)
print("Wrote students.xlsx")
```

```python
# Read from Excel
df2 = pd.read_excel("students.xlsx")
print(df2.head())
# Filter and sort
top = df2[df2["score"] >= 90].sort_values("score", ascending=False)
top
# Export filtered results
top.to_excel("top_students.xlsx", index=False)
print("Wrote top_students.xlsx")
```

```python
#SQLite Database Connectivity (sqlite3)
import sqlite3

# Create/connect to DB
conn = sqlite3.connect("school.db")
cur = conn.cursor()

# Create table
cur.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    score INTEGER NOT NULL
)
""")

conn.commit()
conn.close()

print("Created/verified school.db and students table")
```

```python
import sqlite3

with sqlite3.connect("school.db") as conn:
    cur = conn.cursor()

    # Insert rows (parameterized)
    cur.execute("INSERT INTO students (name, score) VALUES (?, ?)", ("Ritesh", 95))
    cur.execute("INSERT INTO students (name, score) VALUES (?, ?)", ("Nikhil", 88))
    cur.execute("INSERT INTO students (name, score) VALUES (?, ?)", ("Phenome", 91))
    cur.execute("INSERT INTO students (name, score) VALUES (?, ?)", ("Rahul", 77))

print("Inserted sample rows")
```

```python
import sqlite3

with sqlite3.connect("school.db") as conn:
    cur = conn.cursor()

    # SELECT + fetchall
    cur.execute("SELECT name, score FROM students WHERE score >= ? ORDER BY score DESC", (90,))
    rows = cur.fetchall()

print("Students with score >= 90:")
for name, score in rows:
    print(name, score)
```

```python
import sqlite3

with sqlite3.connect("school.db") as conn:
    cur = conn.cursor()

    # UPDATE
    cur.execute("UPDATE students SET score = ? WHERE name = ?", (99, "Ritesh"))

    # DELETE
    cur.execute("DELETE FROM students WHERE name = ?", ("Rahul",))

print("Updated Ruchik, deleted Swapna")
```

```python
import sqlite3

with sqlite3.connect("school.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT id, name, score FROM students ORDER BY id")
    all_rows = cur.fetchall()

print("All rows now:")
for r in all_rows:
    print(r)
```

```python
### Database connectivity and operations using python.
import sqlite3
# connecting with the database.
db = sqlite3.connect("my_database3.db")

# Drop table if it already exist using execute() method.
db.execute("Drop table if exists grades1")

# Create table as per requirement
db.execute("create table grades1(id int, name text, score int)")

#Inserting values inside the created table
db.execute("Insert into grades1(id, name, score) values(101, 'John', 99)")
db.execute("Insert into grades1(id, name, score) values(102, 'Gary', 90)")
db.execute("Insert into grades1(id, name, score) values(103, 'James', 80)")
db.execute("Insert into grades1(id, name, score) values(104, 'Cathy', 85)")
db.execute("Insert into grades1(id, name, score) values(105, 'Anny', 95)")

# commit() some envi auto-commit, but its best practice to do it manually
db.commit()

# close() Ends connection safely
```

```python
results = db.execute(" select * from grades1 order by id")
for row in results:
    print(row)
print("-" * 60)

results = db.execute("select * from grades1 where name = 'John' ")
for row in results: print(row)
print("-" * 60)
db.close()
```
