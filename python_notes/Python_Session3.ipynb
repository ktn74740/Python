Python provides many built-in string methods to manipulate and analyze strings.

Common string operations include:

Accessing characters (indexing & slicing)

Case conversion (upper(), lower(), title())

Searching (find(), count())

Replacing (replace())

Splitting & joining (split(), join())

Checking methods (isalpha(), isdigit(), etc.)

Length calculation (len())

```python
s1 = "Hello"
s2 = 'Python'
s3 = """Multi-line
String"""

print(s1)
print(s2)
print(s3)
```

```python
#Indexing and Slicing
text = "Python"

print(text[0])      # First character
print(text[-1])     # Last character
print(text[0:4])    # Slicing
print(text[::-1])   # Reverse string
```

```python
#Case Conversion Methods
message = "hello world"

print(message.upper())
print(message.lower())
print(message.title())
print(message.capitalize())
```

```python
#Searching Methods
text = "programming in python"

print(text.find("python"))   # Returns index
print(text.count("m"))       # Count occurrences
```

```python
#Basic Slicing
text = "Python"

print(text[0:4])   # Characters from index 0 to 3
print(text[2:])    # From index 2 to end
print(text[:3])    # From start to index 2
```

```python
#Step in Slicing
text = "Python"

print(text[::2])    # Every second character
print(text[1:6:2])  # From index 1 to 5, step 2
```

```python
#Negative Indices
text = "Python"

print(text[-4:])    # Last 4 characters
print(text[:-2])    # All except last 2
```

```python
#Reversing a String
text = "Python"

reversed_text = text[::-1]
print(reversed_text)
```

```python
#String Concatenation
str1 = "Hello"
str2 = "World"

result = str1 + " " + str2
print(result)

#Using join() Method
words = ["Python", "is", "awesome"]

sentence = " ".join(words)  # Join with space
print(sentence)

sentence_dash = "-".join(words)  # Join with dash
print(sentence_dash)


#Using f-strings
name = "Alice"
age = 25

greeting = f"My name is {name} and I am {age} years old."
print(greeting)

#Mixing String with Other Data Types
name = "Charlie"
score = 95

# Using str() to cast integer to string
result = "Student " + name + " scored " + str(score)
print(result)
```

# List Methods

```python
#Adding Elements
fruits = ["apple", "banana"]

fruits.append("cherry")        # Add at the end
fruits.insert(1, "orange")     # Insert at index 1
fruits.extend(["kiwi", "mango"])  # Add multiple elements

print(fruits)
```

```python
#Removing Elements
fruits = ["apple", "orange", "banana", "kiwi", "mango"]

fruits.remove("orange")    # Remove by value
popped = fruits.pop()      # Remove last element
del fruits[1]              # Remove element at index 1
fruits.clear()             # Remove all elements

print(fruits)
print("Popped element:", popped)
```

```python
#Searching Elements
numbers = [10, 20, 30, 20, 40]

print(numbers.index(30))   # Index of first occurrence
print(numbers.count(20))   # Count occurrences of 20
```

```python
#Sorting and Reversing
numbers = [5, 2, 9, 1, 7]

numbers.sort()      # Sort ascending
print("Sorted:", numbers)

numbers.reverse()   # Reverse the list
print("Reversed:", numbers)
```

```python
original = [1, 2, 3]
copy_list = original.copy()

print("Original:", original)
print("Copy:", copy_list)
```
