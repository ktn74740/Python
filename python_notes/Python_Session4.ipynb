Tuples
In Python, tuples are similar to lists but they are immutable i.e. they cannot be changed. You would use the tuples to present data that shouldn't be changed, such as days of week or dates on a calendar.



```python
#Creating Tuples
# Empty tuple
t1 = ()

# Tuple with numbers
t2 = (1, 2, 3)

# Tuple with mixed types
t3 = (1, "apple", True, 3.14)

# Nested tuple
t4 = (1, (26, 3), 4)

print(t2)
print(t3)
print(t4)
```

```python
fruits = ("apple", "banana", "cherry", "date")

print(fruits[0])     # First element
print(fruits[-1])    # Last element
print(fruits[1:3])   # Slicing
```

```python
numbers = (1, 2, 3, 2, 4)

print(numbers.count(2))   # Count occurrences of 2
print(numbers.index(3))   # Index of first occurrence of 3
```

```python
#Iterating Through Tuple
fruits = ("apple", "banana", "cherry")

for fruit in fruits:
    print(fruit)

# Using index
for i in range(len(fruits)):
    print(i, fruits[i])
```

```python
#Tuple Immutability
t = (1, 2, 3)

t[1] = 5   # This will raise an error because tuples are immutable

# You can create a new tuple by concatenation
t2 = t + (4, 5)
print(t2)
```

A set in Python is an unordered, mutable collection of unique elements.

Key features of sets:

Unordered → no indexing

No duplicate values allowed

Mutable → you can add or remove elements

Defined using { } or set() function

Sets are commonly used for:

Removing duplicates

Membership testing

Mathematical set operations (union, intersection, difference)

```python
# Creating a set
s1 = {1, 2, 3, 4}

# Set automatically removes duplicates
s2 = {1, 2, 2, 3, 4, 4}

# Creating set using set() function
s3 = set([5, 6, 7])

print(s1)
print(s2)
print(s3)
```

```python
#Adding Elements
numbers = {1, 2, 3}

numbers.add(4)          # Add single element
numbers.update([5, 6])  # Add multiple elements

print(numbers)
```

```python
#Removing Elements
numbers = {1, 2, 3, 4, 5}

numbers.remove(3)   # Removes element (error if not found)
numbers.discard(10) # No error if element not found
popped = numbers.pop()  # Removes random element

print(numbers)
print("Popped element:", popped)
```

```python
#Set Operations (Mathematical)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

# Union
print(a | b)

# Intersection
print(a & b)

# Difference
print(a - b)

# Symmetric Difference
print(a ^ b)
```

```python
#Membership Testing
fruits = {"apple", "banana", "cherry"}

print("apple" in fruits)
print("grape" not in fruits)
```

```python
#Iterating Through Set
colors = {"red", "green", "blue"}

for color in colors:
    print(color)
```

```python
s = {1, 2, 3}
s.remove(5)
print(s)
```

```python
#Creating a Dictionary
# Empty dictionary
d1 = {}

# Dictionary with values
student = {
    "name": "Alice",
    "age": 22,
    "course": "AI"
}

print(student)
```

```python
#Accessing Values
student = {
    "name": "Alice",
    "age": 22,
    "course": "AI"
}

print(student["name"])        # Access using key
print(student.get("age"))     # Using get() method
```

```python
#Adding and Updating Values
student = {
    "name": "Alice",
    "age": 22
}

# Add new key
student["course"] = "ML"

# Update existing key
student["age"] = 23

print(student)
```

```python
#Removing Items
student = {
    "name": "Alice",
    "age": 22,
    "course": "AI"
}

student.pop("age")      # Remove specific key
del student["course"]   # Delete key
student.clear()         # Remove all items

print(student)
```

```python
#Looping Through Dictionary
student = {
    "name": "Alice",
    "age": 22,
    "course": "AI"
}

for key in student:
    print(key, ":", student[key])

# Using items()
for key, value in student.items():
    print(key, "->", value)
```

```python
nested = {"key1": {"nestkey": {"subnestkey": 123}}}
nested["key1"]["nestkey"]["subnestkey"]
```

```python
# Dictionary Comprehensions
# Just like list comprehensions, you can build dictionaries quickly.
# Example: squares from 1..10:

{x: x**2 for x in range(1, 11)}
squares = {x: x**2 for x in range(1, 11)}
squares
```

# Functions
A function is reusable code.

Why functions?

Avoid repeating code
Make code organized and readable
Makes testing and debugging easier
Syntax:

def function_name(arg1, arg2):
    """docstring"""
    # body
    return value

```python
def print_num(n):
    """Print multiples of n from 1*n to n*n."""
    for i in range(1, n + 1):
        print(i * n)

print_num(8)
```

```python
def func_with_print():
    print("Hello")

def func_with_return():
    return "Hello"

a = func_with_print()
b = func_with_return()

print("a =", a)  # None
print("b =", b)  # Hello
```

```python
def is_prime(num):
    """Naive prime check. Prints 'prime' or 'not prime'."""
    if num < 2:
        print("not prime")
        return

    for n in range(2, num):
        if num % n == 0:
            print("not prime")
            break
    else:
        print("prime")

is_prime(9)
is_prime(13)
is_prime(1)
```
